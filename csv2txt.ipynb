```python
import os
import shutil
from google.colab import files

# === Step 1: Set up directories ===
input_path = "input"
output_path = "output"
os.makedirs(input_path, exist_ok=True)
os.makedirs(output_path, exist_ok=True)

# === Step 2: Upload files ===
print("🔼 Please upload your input files:")
uploaded = files.upload()  # User uploads files

# === Step 3: Move uploaded files to input_path ===
for filename in uploaded:
    os.rename(filename, os.path.join(input_path, filename))

# === Step 4: Process each file ===
for data in os.listdir(input_path):
    if data.startswith(".") or data.startswith("..") or len(data) < 12:
        continue

    full_input_path = os.path.join(input_path, data)
    output_file_name = f"{os.path.splitext(data)[0]}.txt"
    full_output_path = os.path.join(output_path, output_file_name)

    with open(full_input_path, encoding='utf-8-sig') as input_file:
        with open(full_output_path, "w") as output_file:
            lines = input_file.readlines()
            for line in lines:
                line = line.strip()
                chk = line.split(",")
                if len(chk) >= 2:
                    column1 = chk[0]
                    column2 = chk[1]
                    output_file.write(column1 + ' ' + column2 + '\n')
                    print(column1, column2)

# === Step 5: Zip and download all output files ===
#It'll download it to your local downloads directory
shutil.make_archive("processed_output", 'zip', output_path)
print("⬇️ Downloading processed_output.zip...")
files.download("processed_output.zip")
```
