```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless=new")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 20)
```

```python
import time
import getpass

# üîê Prompt for credentials securely
TILEDB_EMAIL = input("Enter your TileDB login email: ")
TILEDB_PASSWORD = getpass.getpass("Enter your TileDB password: ")



# Go to the login page
driver.get("https://www.particle.cz/work/TileDB2/account/login")

# Wait for the email field to be present
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "MainContent_loginEmailTxt"))
)

# Fill in credentials and log in
driver.find_element(By.ID, "MainContent_loginEmailTxt").send_keys(TILEDB_EMAIL)
driver.find_element(By.ID, "MainContent_loginPasswordTxt").send_keys(TILEDB_PASSWORD)
driver.find_element(By.ID, "MainContent_LoginBtn").click()

# Give time for login to complete
time.sleep(3)
```

```python
import os
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

UPLOAD_FOLDER = "content/Burn_In" #!!!!!!!!!! <<<<<<<<<<<<< CHANGE FILE LOCATION HERE!!!!!!!!!!!!!!!!!
uploaded_count = 0

files = [f for f in os.listdir(UPLOAD_FOLDER) if f.endswith(".TXT")]
print(f"üìÇ Files in {UPLOAD_FOLDER}: {files}\n")

for file_name in files:
    file_path = os.path.join(UPLOAD_FOLDER, file_name)
    brick_code = file_name.split('_')[0].strip().upper()



    print(f"üîç Checking: {file_name} -> Brick Code: {brick_code}")
    measure_url = f"https://www.particle.cz/work/TileDB2/brick10v/measure?partcode={brick_code}"
    print(f"üåê Navigating to: {measure_url}")
    driver.get(measure_url)

    try:
        wait = WebDriverWait(driver, 10)

        # Upload file
        upload_input = wait.until(EC.presence_of_element_located((By.ID, "MainContent_measureFileUpl")))
        upload_input.send_keys(os.path.abspath(file_path))
        print(f"üì§ File attached: {file_name}")

        # Fill Note
        note_input = driver.find_element(By.ID, "MainContent_measureNoteTxt")
        note_input.clear()
        note_input.send_keys("Completed Burn-In Testing")

        # Fill Author
        author_input = driver.find_element(By.ID, "MainContent_measureAuthorCB_measureAuthorCB_TextBox")
        author_input.clear()
        author_input.send_keys("Thabo Pilusa")
        # üí° Click the Measure Date field to dismiss the dropdown
        driver.find_element(By.ID, "MainContent_brickMeasuredTxt").click()
        time.sleep(0.5)




        # Submit
        submit_button = driver.find_element(By.ID, "MainContent_brickMeasureBtn")
        submit_button.click()
        print(f"‚úÖ Successfully uploaded diary entry for {brick_code}\n")
        uploaded_count += 1
        time.sleep(2)

    except Exception as e:
        print(f"‚ùå Failed to upload {file_name}:\n{e}\n")

print(f"üöÄ Uploading completed. Total files uploaded: {uploaded_count}")
```
